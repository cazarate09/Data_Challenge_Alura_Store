# **2. Ventas por categoría**



---



```python
def formatear_valor_ventas(valor)->str:
    return f"${valor:,.2f}"
```

## 2.1. Ventas por Categoría Tienda Uno

```python
ventas_categoria_tienda_uno = tienda['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda_uno.columns = ['Categoría del Producto', 'Cantidad Ventas']
precio_categoria_tienda_uno = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
precio_categoria_tienda_uno.columns = ['Categoría del Producto', 'Precio']
ventas_categoria_tienda_uno = ventas_categoria_tienda_uno.merge(precio_categoria_tienda_uno, on='Categoría del Producto')
ventas_categoria_tienda_uno['Precio'] = ventas_categoria_tienda_uno['Precio'].apply(formatear_valor_ventas)
ventas_categoria_tienda_uno

#ventas_categoria_tienda_uno
```

## 2.2. Ventas por Categoría Tienda Dos

```python
ventas_categoria_tienda_dos = tienda2['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda_dos.columns = ['Categoría del Producto', 'Cantidad Ventas']
precio_categoria_tienda_dos = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
precio_categoria_tienda_dos.columns = ['Categoría del Producto', 'Precio']
ventas_categoria_tienda_dos = ventas_categoria_tienda_dos.merge(precio_categoria_tienda_dos, on='Categoría del Producto')
ventas_categoria_tienda_dos['Precio'] = ventas_categoria_tienda_dos['Precio'].apply(formatear_valor_ventas)
ventas_categoria_tienda_dos
```

## 2.3. Ventas por Categoría Tienda Tres

```python
ventas_categoria_tienda_tres = tienda3['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda_tres.columns = ['Categoría del Producto', 'Cantidad Ventas']
precio_categoria_tienda_tres = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
precio_categoria_tienda_tres.columns = ['Categoría del Producto', 'Precio']
ventas_categoria_tienda_tres = ventas_categoria_tienda_tres.merge(precio_categoria_tienda_tres, on='Categoría del Producto')
ventas_categoria_tienda_tres['Precio'] = ventas_categoria_tienda_tres['Precio'].apply(formatear_valor_ventas)
ventas_categoria_tienda_tres
```

## 2.4. Ventas por Categoría Tienda Cuatro

```python
ventas_categoria_tienda_cuatro = tienda4['Categoría del Producto'].value_counts().reset_index()
ventas_categoria_tienda_cuatro.columns = ['Categoría del Producto', 'Cantidad Ventas']
precio_categoria_tienda_cuatro = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()
precio_categoria_tienda_cuatro.columns = ['Categoría del Producto', 'Precio']
ventas_categoria_tienda_cuatro = ventas_categoria_tienda_cuatro.merge(precio_categoria_tienda_cuatro, on='Categoría del Producto')
ventas_categoria_tienda_cuatro['Precio'] = ventas_categoria_tienda_cuatro['Precio'].apply(formatear_valor_ventas)
ventas_categoria_tienda_cuatro
```

```python

tiendas = ('Tienda_1', 'Tienda_2', 'Tienda_3', 'Tienda_4')

Facturacion = [ingresos_tienda_1, ingresos_tienda_2, ingresos_tienda_3, ingresos_tienda_4]
valor_total_envios = [valor_total_envios_tienda_1, valor_total_envios_tienda_2, valor_total_envios_tienda_3, valor_total_envios_tienda_4]
valor_promedio_envio = [valor_promedio_envio_tienda_1, valor_promedio_envio_tienda_dos, valor_promedio_envio_tienda_tres, valor_promedio_envio_tienda_cuatro]

x = np.arange(len(tiendas))
#Ancho de las barras
width = 0.25

fig, ax = plt.subplots(figsize=(10, 8))

barra_1 = ax.bar(x - width, Facturacion, width, label='Facturación', color = 'black')
barra_2 = ax.bar(x, valor_total_envios, width, label='Valor Total de Envios', color = 'blue')
barra_3 = ax.bar(x + width, valor_promedio_envio, width, label='Valor Promedio de Envío', color = 'red')

for barra in (barra_1, barra_2, barra_3):
  for rectangulo in barra:
    altura = rectangulo.get_height()
    ax.text(rectangulo.get_x() + rectangulo.get_width() / 2, altura, f"${altura:,.2f}", ha='center', va='bottom', fontsize = 7)

# Personalizar el gráfico
ax.set_xlabel("Tiendas")
ax.set_ylabel("Valores")
ax.set_title("Comparación de Facturación y Costos de Envío por Tienda")
ax.set_xticks(x)
ax.set_xticklabels(tiendas)
ax.set_ylim(0, max(max(Facturacion), max(valor_total_envios), max(valor_promedio_envio)) * 1.2)  # Ajusta la altura
ax.legend()

# Mostrar el gráfico
plt.show()
```
